```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2
```

# original data

```python
tr_path = '../../data/processed/training-set.csv'
```

```python
df = pd.read_csv(tr_path)
        
x = df.drop(['id', 'attack_cat', 'label'], axis=1)
y = pd.Categorical(df['attack_cat']).codes.astype(np.float64)
```

```python
x.shape
```

```python
# 資料的欄位名稱
cols = x.columns

colNum = cols.to_numpy() 
colSet = set(cols.to_numpy())
```

```python
colSet
```

# feature selection using Chi-Square Test

https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectPercentile.html
https://medium.com/jackys-blog/%E6%80%8E%E9%BA%BC%E5%81%9A%E4%B8%80%E5%80%8B%E5%A5%BD%E7%9A%84%E7%89%B9%E5%BE%B5%E9%81%B8%E6%93%87-%E4%BD%BF%E7%94%A8python-part-1-d698cc15d76e

```python
selected = SelectPercentile(chi2,percentile=30)
```

```python
#只選得分最好的前30%features
x_new=selected.fit_transform(x,y)
```

```python
x_new.shape
```

```python
# 目前選出前30%有用的特徵
selectedColumn = selected.get_feature_names_out()

selectedColumnSet = set(selectedColumn)
```

```python
selectedColumnSet
```

```python
removeColumnSet = colSet - selectedColumnSet

removeColumnSet
```
